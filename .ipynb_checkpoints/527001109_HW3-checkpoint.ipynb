## 1.1 Load the Data

Please download the dataset from Piazza. There are about 65,000 ratings in total. We split the rating data into two set. You will train with 70% of the data (in *train_movie.csv*) and test on the remaining 30% of data (in *test_movie.csv*). Each of train and test files has lines having this format: UserID, MovieID, Rating. 

First you will need to load the data and store it with any structure you like. Please report the numbers of unique users and movies in the dataset. 

```python
# load the data, then print out the number of
# movies and users in each of train and test sets.
# Your Code Here...
import csv
def read_file(file):
    f = open(file)
    data=[]
    csv_f = csv.reader(f)
    first = True
    for line in csv_f:
        if(first==False):
            data.append(line)
        first = False
    return data

train_movie= read_file('train_movie.csv')
test_movie= read_file('test_movie.csv')

def configure_data(data):
    user_data={}
    movie_set=set()
    for row in data:
        user_id= float(row[0])
        movie_id= float(row[1])
        rating=float(row[2])
        if user_id not in user_data:
            user_data[user_id]={}
        user_data[user_id][movie_id]=rating
        movie_set.add(movie_id) 
    return user_data, movie_set

user_ratings_train, movie_set_train = configure_data(train_movie)
user_ratings_test, movie_set_test = configure_data(test_movie)

print("Number of users in train set is "+ repr(len(user_ratings_train)))
print("Number of movies in train set is "+  repr(len(movie_set_train)))
print("Number of users in test set is "+ repr(len(user_ratings_test)))
print("Number of movies in test set is "+  repr(len(movie_set_test)))
```

### Cosine Similarity

First, you will try to calculate the similarity between users with cosine similary following the equation on page 16 of [Collaborative Filtering Recommender Systems](http://files.grouplens.org/papers/FnT%20CF%20Recsys%20Survey.pdf). And then you need to predict the rating for each (user, movie) tuple in the test set. *Note: you don't need to subtract user mean baseline from the ratings prior to computing the similarity.*

```python
avg_rating={}
for user in user_ratings_train.keys():
    avg_rating[user]= sum(user_ratings_train[user].values())/ float(len(user_ratings_train[user]))
```

```python
import math
norm_scores={}
for user in user_ratings_train.keys():
    nscore=0
    for val in user_ratings_train[user].values():
        nscore += val**2
    norm_scores[user]= math.sqrt(nscore)
#norm_scores
```

```python
def find_cos_score(u1, u2):
    score=0
    for mov in user_ratings_train[u1].keys():
        if(mov in user_ratings_train[u2]):
            score+= user_ratings_train[u1][mov]+ user_ratings_train[u2][mov]
    return score/(norm_scores[u1]* norm_scores[u2]) 
```

```python
# Your Code Here...
# Predict for test set



def preidct_score(evalparam, k):
    predicted=[]
    count=0
    for row in test_movie:
        uid = float(row[0])
        movie= float(row[1])
        score= avg_rating[uid]
        inscore=0
        #print("user1 is "+ repr(uid))
        for key in user_ratings_train.keys():
            user_rat=0
            if(movie  in user_ratings_train[key]):
                user_rat=user_ratings_train[key][movie]
                pairscore=0
                if evalparam=='C':
                    pairscore= (user_rat-avg_rating[key])*find_cos_score(uid, key)
                elif (evalparam=='P'):
                    pairscore= (user_rat-avg_rating[key])*find_pearson_score(uid, key)
                inscore += pairscore
        score+= k*inscore
        count+=1
        print("score for row  "+ repr(count) + " is "+ repr(score))  
        predicted.append(score)
    return predicted
    
```

```python
ground_truth=[]
for row in test_movie:
    ground_truth.append(float(row[2]))
```

```python
def find_RMS(list1, list2):
    rmscore=0
    for i in range(0,len(list1)):
        rmscore += (list1[i]- list2[i])**2
    return math.sqrt(rmscore/len(list1))

def find_ABS(list1, list2):
    abscore=0
    for i in range(0,len(list1)):
        abscore += abs(list1[i]- list2[i])
    return (abscore/len(list1))
```

### Pearson correlation

Then, you will try to calculate the similarity between users with pearson correlation following `week06rec.pdf` (Page 37). And then you need to predict the rating for each (user, movie) tuple in the test set.

```python
# Your Code Here...
# Predict for test set
def find_pearson_score(u1, u2):
    avg1= avg_rating[u1]
    avg2=  avg_rating[u2]
    num=0
    den1=0
    den2=0
    for mov in user_ratings_train[u1].keys():
        if mov in user_ratings_train[u2]:
            num+= (user_ratings_train[u1][mov]-avg1)*(user_ratings_train[u2][mov]-avg2)
            den1+= (user_ratings_train[u1][mov]-avg1)**2
            den2+= (user_ratings_train[u2][mov]-avg2)**2
            
    if(den1==0 or den2==0):
        return 5
    den1= math.sqrt(den1)
    den2= math.sqrt(den2)
    return num/(den1*den2)  
```
